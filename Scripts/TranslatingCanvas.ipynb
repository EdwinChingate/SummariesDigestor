```python
import json

# Load the .canvas file
file_path = "your_canvas_file.canvas"  # Replace with your file path
with open(CanvasName, "r", encoding="utf-8") as file:
    canvas_data = json.load(file)

# Extract nodes (cards)
nodes = {node["id"]: node for node in canvas_data.get("nodes", [])}

# Extract connections (edges)
edges = canvas_data.get("edges", [])

# Display summary
print(f"Total Nodes: {len(nodes)}")
print(f"Total Connections: {len(edges)}")

# Example: Print first node content
if nodes:
    first_node_id, first_node = next(iter(nodes.items()))
    print(f"\nFirst Node (ID: {first_node_id}):")
    print(json.dumps(first_node, indent=4))
```

```python
NodesKeys=nodes.keys()
TextList=[]
Loc='0-Vault/01-Projects/Repository-feat.-ms2-Gauss/Documentation/Capabilities/WorkflowStory/Elements of my story/'
for key in NodesKeys:
    node=nodes[key]
    x=node['x']
    y=node['y']
    NodeType=node['type']
    if NodeType=='file':
        Sentence=str(node['file'])
        Sentence=Sentence.replace(Loc,'')
        Sentence=Sentence[12:]
        #print(Sentence.find('HRMS_WF-'))
        TextList.append([Sentence,x,y])
```

```python
NodesKeys=nodes.keys()
ParagraphsList=[]
for key in NodesKeys:
    node=nodes[key]
    x=node['x']
    y=node['y']
    NodeType=node['type']
    if NodeType=='group':
        Sentence="\n"#node['label']
        height=node['height']
        ParagraphsList.append([Sentence,x,y])
        ParagraphsList.append([Sentence,x,y+height])
```

```python
TextDF=pd.DataFrame(TextList,columns=['Sentence','x','y'])
ParagraphsDF=pd.DataFrame(ParagraphsList,columns=['Sentence','x','y'])
```

```python
DraftDF=pd.concat([TextDF,ParagraphsDF])
DraftDF=DraftDF.sort_values(by='y')
```
