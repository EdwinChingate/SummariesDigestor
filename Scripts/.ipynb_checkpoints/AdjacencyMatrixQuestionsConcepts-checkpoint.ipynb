```python
QuestionsConceptsArray=np.array(QuestionsConceptsDF)
```

```python
#This function returns a list with the ids of the interacting questions.
def NetworkList_from_Array(ids_questions_with_concept,RelatedQuestionsList):
    NQuestions=len(ids_questions_with_concept)
    for Locquestion_1 in np.arange(NQuestions):
        question_1=ids_questions_with_concept[Locquestion_1]
        for Locquestion_2 in np.arange(Locquestion_1+1,NQuestions):
            question_2=ids_questions_with_concept[Locquestion_2]
            edge=[question_1,question_2]
            RelatedQuestionsList.append(edge)
    return RelatedQuestionsList
```

```python
def AdjacencyMatrixQuestions(ids_questions_with_concept,QAdjacencyMatrix):
    NQuestions=len(ids_questions_with_concept)
    for Locquestion_1 in np.arange(NQuestions):
        question_1=ids_questions_with_concept[Locquestion_1]
        for Locquestion_2 in np.arange(Locquestion_1+1,NQuestions):
            question_2=ids_questions_with_concept[Locquestion_2]
            QAdjacencyMatrix[question_1,question_2]+=1
            #QAdjacencyMatrix[question_2,question_1]+=1
    return QAdjacencyMatrix
```

```python
NConceptColumns=QuestionsConceptsArray.shape[1]
NQuestionRows=QuestionsConceptsArray.shape[0]
QAdjacencyMatrix=np.zeros((NQuestionRows,NQuestionRows))
RelatedQuestionsList=[]
for conceptColumn in np.arange(NConceptColumns):
    ids_questions_with_concept=np.where(QuestionsConceptsArray[:,conceptColumn]>0)[0]
    #RelatedQuestionsList=NetworkList_from_Array(ids_questions_with_concept=ids_questions_with_concept,RelatedQuestionsList=RelatedQuestionsList)
    QAdjacencyMatrix=AdjacencyMatrixQuestions(ids_questions_with_concept=ids_questions_with_concept,QAdjacencyMatrix=QAdjacencyMatrix)
    #print(ids_questions_with_concept)
```

```python
import networkx as nx
```

```python
GraphQ=nx.from_numpy_array(QAdjacencyMatrix)
```

```python
#This is the exact algoritm
bc = nx.betweenness_centrality(
         GraphQ,
         weight="none",      # ← or None for unweighted
         normalized=True)      # in [0,1]
```

```python
#This is an approximation
bc = nx.betweenness_centrality(
         GraphQ,
         k=300,                # sample 300 source nodes
         weight="none",
         normalized=True,
         seed=42)              # reproducible
```

```python
bc
```

```python
# Convert to a sorted Series and save
pd.Series(bc, name="betweenness").sort_values(ascending=False) \
        .to_csv("betweenness_scores.csv")

# Quick peek
print(pd.Series(bc).nlargest(10))
```
